## Karakteristik Data

### Data yang kami ambil mengandung 11 atribut, yakni:

1. work_year            : Berisi data waktu dalam satuan tahun (Categorical-Nominal)
2. experience_level     : Berisi data pengalaman kerja (Categorical-Nominal)
3. employment_type      : Berisi data jenis pekerjaan (Categorical-Nominal)
4. job_title            : Berisi data nama pekerjaan (Categorical-Nominal)
5. salary               : Berisi pendapatan mata uang asal dalam masing-masing mata uang (Quantitative-Discrete)
6. salary_currency      : Berisi nilai mata uang (Categorical-Nominal)
7. salary_in_usd        : Berisi pendapatan dalam mata uang USD (Quantitative-Discrete)
8. employee_residence   : Berisi daerah tempat tinggal pekerja (Categorical-Nominal)
9. remote_ratio         : Berisi persentase dari tingkat 'remote' pekerjaan (Quantitative-Continuous)
10. company_location    : Berisi lokasi perusahaan (Categorical-Nominal)
11. company_size        : Berisi ukuran perusahaan (Categorical-Nominal)







```
 Catatan: (categorical describe)
 1. count = jumlah atribut
 2. unique = jumlah atribut nonduplikat
 3. top = nilai yang paling sering muncul
 4. freq = jumlah dari nilai yang paling sering muncul
```


```
 Catatan: (quantitative describe)
 1. count = jumlah atribut
 2. mean = nilai rata-rata
 3. std = nilai standar deviasi
 4. min = nilai terkecil
 5. 25 % = quatil pertama
 6. 50 % = median
 7. 75 % = kuartil ketiga
 8. max = nilai terbesar
```

### Statistik

```
 Informasi / pengetahuan yang didapatkan:
 1. Mean (rata-rata) = Ukuran pemusatan dari suatu sebaran probabilitas
 2. Trimmed Mean = mereduksi nilai rata-rata yang terlalu menyimpang karena nilai ekstremum
 3. Median (nilai tengah) = nilai yang berada di tengah data
 4. Modus (nilai yang sering muncul) = nilai yang sering muncul
 5. Variance (varians) = menggambarkan varians kuantitas acak sebagai fungsi meannya
 6. Standar Deviasi = menentukan seberapa dekat data dari sampel statistik dengan data rata-rata data tersebut
 7. Mean Absolute Deviation = mengukur kesalahan perkiraan dalam unit ukuran yang sama seperti data aslinya
 8. Percentile = nilai yang berada pada suatu bagian yang terbagi menjadi beberapa partisi persentil
 9. Ekstremum = nilai terbesar dan nilai terkecil
 10. Range (jangkauan) = selisih nilai terbesar dan nilai terkecil
 11. Interquartile Range = Selisih persentil ke-75 dengan persentil ke-25
 12. Outlier = penyimpangan jauh dari beberapa data
```

#### Metode .fillna(value = _)

```
 Metode ini cocok untuk membersihkan data pada kolom bertipe quantitave discrete maupun quantitative continuous, karena dengan metode ini, data yang bernilai null ('-', NaN, dll) akan diubah menjadi suatu value yang diinginkan
```



````python
# data yang bernilai null ('-', NaN, dll) akan diubah menjadi value yang diisi, dalam hal ini bernilai 0

dfna = df.fillna(value = 0) # value dapat bernilai integer atau float
dfna
````

#### Metode .fillna(method = 'pad')

```
 Metode ini cocok untuk membersihkan data pada kolom bertipe categorical, karena dengan metode ini, data yang bernilai null ('-', NaN, dll) akan diubah menjadi value yang sama pada baris sebelumnya. Jangan gunakan method ini jika kolom bersifat unique
```

````python
# data yang bernilai null akan diubah menjadi nilai data yang ada pada baris sebelumnya (baris di atasnya)

dfnaa = df.fillna(method = 'pad') # gunakan axes='' setelah method='pad' jika diperlukan untuk mengisi value dengan menggeser sesuai nilai axes
dfnaa
````

#### Metode .fillna(method = 'bfill')

```
 Metode ini cocok untuk membersihkan data pada kolom bertipe categorical, karena dengan metode ini, data yang bernilai null ('-', NaN, dll) akan diubah menjadi value yang sama pada baris setelahnya. Jangan gunakan method ini jika kolom bersifat unique
```

````python
# data yang bernilai null akan diubah menjadi data yang ada pada baris setelahnya (baris di bawahnya)

dfb = df.fillna(method = 'bfill') # gunakan axes='' setelah method='bfill' jika diperlukan untuk mengisi value dengan menggeser sesuai nilai axes
dfb
````

#### Mengisi null values dengan mengisi values (dengan method .fillna( ))

```
 Metode ini cocok untuk membersihkan data pada kolom bertipe categorical maupun quantitative, karena dengan metode ini, data yang bernilai null ('-', NaN, dll) akan diubah sesuai dengan input
```
